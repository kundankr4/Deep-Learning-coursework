(1) Load Salary_data.csv

```python
#import matplotlib
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("Salary_Data.csv",delimiter=',', skiprows=1)
print(data[4])
```

(2) Plot data

```python
years = data[:, 0]
salary = data[:, 1]

plt.plot(years, salary)
plt.xlabel('YearsExperiences')
plt.ylabel('Salary')
plt.title('Experiences vs. Salary')
plt.grid()
```
